Благодаря использованию Grafana появляется возможность создавать красивые и понятные дашборды, отображающие в реальном времени различные метрики и данные.

Благодаря Grafana гораздо проще проводить анализ данных.

<b>Выводы по графикам</b>
1. Исходя из графика распределения значений по диапазонам, наибольшее количество значений находится в диапазоне от 900 до 1000.
2. Распределение общей стоимости по категориям примерно одинаково для категорий С, D и А. Наименьшее значение у категории В.
3. Изменение общей стоимости по времени имеет стохастический характер. Закономерность не наблюдается, скорее всего ввиду того, что данные вымышленные.
4. Общая стоимость в разных регионах находится примерно в одном диапазоне. Наибольшее значение имеет западный регион. Наименьшее - северный.
5. Доля продуктов от общей стоимости находится примерно в одном диапазоне. Наибольшее значение у продукта В, наименьшее - у продукта С.
