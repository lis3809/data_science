```python
from faker import Faker
import random
import datetime

fake = Faker()

# Generate and print the CREATE TABLE statement
create_table_sql = """
CREATE TABLE grafana_data (
    id serial PRIMARY KEY,
    timestamp timestamp,
    category varchar(50),
    value float,
    region varchar(50),
    department varchar(50),
    product varchar(50),
    status varchar(20)
);
"""

print(create_table_sql)

# Generate and print the INSERT INTO statements
insert_statements = []
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
regions = ['North', 'South', 'East', 'West']
departments = ['Sales', 'Marketing', 'Finance', 'Operations']
products = ['Product A', 'Product B', 'Product C', 'Product D']
statuses = ['Pending', 'Processing', 'Completed']

for _ in range(200):
    timestamp = fake.date_time_this_year()
    category = random.choice(categories)
    value = round(random.uniform(10, 1000), 2)
    region = random.choice(regions)
    department = random.choice(departments)
    product = random.choice(products)
    status = random.choice(statuses)

    insert_statement = f"INSERT INTO grafana_data (timestamp, category, value, region, department, product, status) VALUES ('{timestamp}', '{category}', {value}, '{region}', '{department}', '{product}', '{status}');"
    insert_statements.append(insert_statement)

for statement in insert_statements:
    print(statement)
```
