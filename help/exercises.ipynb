```python
def func(m):
    return (200 * (6 - m)**2)
```

```python
def gradient(m):
    return -400 * (6 - m)
```

```python
def gradient_descent(n_interations, step):
    m_0 = 1
    
    m_old = m_0
    ms = [m_old]
    for i in range(n_interations):
        m_new = m_old - step * gradient(m_old)
        m_old = m_new
        ms.append(m_old)

    return m_old, ms
```

```python
result, results_all = gradient_descent(10, 0.001)
plt.figure(figsize=(18, 6))
pd.Series([func(m) for m in range(1, 13)], index=range(1, 13)).plot()
pd.Series([func(x) for x in results_all], index = results_all).plot(marker='o')
```
